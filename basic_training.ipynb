```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

```python
X_train_tensor.sha
```

```python
from torch import nn as nn
```

define the model

```python
class MyNetwork(nn.Module):
  def __init__(self,num_features,out_features) :
    super().__init__()
    self.model = nn.Sequential(
        nn.Linear(num_features,4),
        nn.ReLU(),
        nn.Linear(4,1),
        nn.Sigmoid()
    )
  def forward(self,features):
    out=self.model(features)
    return out
```

```python
num_features= X_train_tensor.shape[1]
out_features = 1
```

```python
learning_rate=0.1
epoch=50
```

```python
import torch.optim as optim
#define model
model=MyNetwork(num_features,out_features )

#define loss
loss_function= nn.BCELoss()

#define optimizer
optimizer = optim.Adam(model.parameters(), lr=0.01)
#define loop

for epoch in range(epoch):

  #forward pass
  y_pred = model(X_train_tensor.float())

  #calculate loss
  loss=loss_function(y_pred,y_train_tensor.view(-1,1))

  # Backward pass & optimization
  optimizer.zero_grad()  # Reset gradients from last step
  loss.backward()        # Calculate new gradients
  optimizer.step()       # Update weights






```

```python
with torch.no_grad():
  y_pred= model.forward(X_test_tensor)
  y_pred = (y_pred>0.7).float()
  acc= (y_pred == y_test_tensor).float().mean()
  print(acc)
```
